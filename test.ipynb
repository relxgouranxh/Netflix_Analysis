WHy do we check null values?

Because by checking null value we are able to sort the data means we can drop the larger number of null values column from the dataset.
eg. director column have 2634 null values which are higher not that much because we have 8807 rows , so we can drop it if we want


## Handling the null values

* Director , country and cast can be replace by - Unknown 
* date can be replace by - median value
* rating and duration  can be replace by - Not rated and none


```python
df.fillna({"director": "Unknown","cast":"Unknown","country":"Unknown"}, inplace=True)
df.director.isnull().sum(),df.cast.isnull().sum(),df.country.isnull().sum()
```

```python
df["date_added"]=pd.to_datetime(df["date_added"],format="%B %d, %Y")
df.fillna({"date_added":df["date_added"].median()},inplace=True)
df.duration.isnull().sum()
```

```python
df.fillna({"rating":"Not rated"},inplace=True)
df.loc[df["rating"]=="Not rated"]
df["rating"].isnull().sum()
```

```python
df.fillna({"duration":"none"},inplace=True)
df.duration.isnull().sum()
```

Checking is still there null values

```python
df.isnull().sum()
```

## Date Added Analysis

As we know that in dataframe the date column have string datatype so we have to convert it into date_time datatype

```python
df["date_added"]=pd.to_datetime(df["date_added"],errors="coerce")
```

Extracting the year from date

```python
df["year"]=df["date_added"].dt.year
```

### Which year highest content added

#### For Movies

```python
#Only Movies
on_movie["date_added"]=pd.to_datetime(on_movie["date_added"],errors="coerce")
on_movie["year"]=on_movie["date_added"].dt.year
top10_yr_mov=on_movie.groupby("year")["type"].value_counts().reset_index()[["year","count"]].sort_values(by="year",ascending=False).head(10)
top10_yr_mov
```

```python
fig=px.bar(top10_yr_mov,x="year",y="count",title="Top 10 years having Highest content added")
fig
```

#### For Tvshows

```python
onli_show["date_added"]=pd.to_datetime(onli_show["date_added"],errors="coerce")
onli_show["year"]=onli_show["date_added"].dt.year
top10_yr_show=onli_show.groupby("year")["type"].value_counts().reset_index()[["year","count"]].sort_values(by="count",ascending=False)
top10_yr_show
```

```python
fig=px.bar(top10_yr_show,x="year",y="count")
fig
```

#### Month wise analysis

Enter the type and year for which you want the data. This will display the month-wise number of content for that particular year and type.        

```python
sel_year=int(input("Enter year : " ))
sel_type=input("Enter the type (Movie or TV Show): ")

df_sel=df.loc[(df["date_added"].dt.year==sel_year)&(df["type"]==sel_type)]
df_sel["date_added"]=pd.to_datetime(df_sel["date_added"])

#getting the month
df_sel["month"]=df_sel["date_added"].dt.month_name()
sel_data_of_year=df_sel.groupby("month")["type"].value_counts().reset_index()[["month","count"]]
sel_data_of_year
```

```python
fig=px.bar(sel_data_of_year,x="month",y="count",title=f"Content added in {sel_year} of type {sel_type} according the months")
fig
```
